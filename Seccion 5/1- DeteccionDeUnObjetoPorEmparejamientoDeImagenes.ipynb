```python
imagen = cv2.imread('perro.jpg')
```

```python
imagen = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB)
plt.imshow(imagen)
```

```python
imagen_2 = cv2.imread('cara.jpg')
imagen_2 = cv2.cvtColor(imagen_2,cv2.COLOR_BGR2RGB)
plt.imshow(imagen_2)
```

```python
metodo = cv2.TM_CCOEFF
resultado = cv2.matchTemplate(imagen,imagen_2,metodo)
plt.imshow(resultado)
```

```python
valor_min, valor_max, pos_min, pos_max = cv2.minMaxLoc(resultado) #Devuelve los valores minimos,maximos y posiciones
```

```python
imagen_2.shape
```

```python
alto, ancho, colores = imagen_2.shape
```

```python
top_izquierda = pos_max
botton_derecha = (pos_max[0]+ancho, pos_max[1]+alto)
```

```python
cv2.rectangle(imagen, top_izquierda,botton_derecha,color=(255,0,0),thickness=8)
plt.imshow(imagen)
```
