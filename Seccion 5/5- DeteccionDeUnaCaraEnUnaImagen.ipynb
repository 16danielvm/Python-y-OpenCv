```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
cd "D:/Universidad/Cursos/Python y OpenCv"
```

```python
rostro1 = cv2.imread('rostro1.png',0)
plt.imshow(rostro1,cmap='gray')
```

```python
familia = cv2.imread('familia.png',0)
plt.imshow(familia,cmap='gray')
```

```python
cascada_cara = cv2.CascadeClassifier('D:/Universidad/Vision por Computador/opencv/build/etc/haarcascades/haarcascade_frontalface_default.xml')
```

```python
def detectar_cara(imagen):
    imagen1 = imagen.copy()
    rectangulos = cascada_cara.detectMultiScale(imagen1)
    for (x,y,w,h) in rectangulos:
        cv2.rectangle(imagen1,(x,y),(x+w,h+y),(255,0,0),10)
    return imagen1
```

```python
resultado = detectar_cara(rostro1) 
plt.imshow(resultado,cmap='gray')
```

```python
resultado = detectar_cara(familia)
plt.imshow(resultado,cmap='gray')
```

```python
resultado = detectar_ojos(rostro1)
plt.imshow(resultado,cmap='gray')
```
