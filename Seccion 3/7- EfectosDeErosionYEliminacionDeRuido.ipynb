```python
#Operadores morfologicos (efectos de erosion, eliminacion del ruido)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
imagen = np.zeros((300,600)) #Creacion de imagen negra de 300 de alto por 600 de ancho
fuente = cv2.FONT_HERSHEY_SIMPLEX
#Crea un texto(ABCDE) que empieza desde el punto (50,200 contando desde abajo e izquierda) color blanco y grosor 8 con dicha fuente
cv2.putText(imagen, text='ABCDE', org=(50,200),fontFace = fuente, fontScale = 5,color=(255,255,255),thickness=8)
plt.imshow(imagen,cmap='gray')
```

```python
nucleo = np.ones((5,5),dtype=np.int8) #Array de 5x5 de unos
```

```python
imagen_2 = cv2.erode(imagen,nucleo,iterations=1) #Erosiona la imagen, recortar letras para que se vean mas estrechas
plt.imshow(imagen_2,cmap='gray')
```

```python
ruido = np.random.randint(low = 0, high=2, size=(300,600)) #Creacion de ruido
plt.imshow(ruido, cmap='gray')
```

```python
#Agregar ruido a la imagen de texto
ruido = ruido * 255
imagen_ruido = imagen + ruido
plt.imshow(imagen_ruido,cmap='gray')
```

```python
#Eliminacion de ruido
imagen_sin_ruido = cv2.morphologyEx(imagen_ruido,cv2.MORPH_OPEN,nucleo) #Funcion para eliminar ruido
plt.imshow(imagen_sin_ruido,cmap='gray')
```

```python
gradiente = cv2.morphologyEx(imagen,cv2.MORPH_GRADIENT,nucleo) #Elimina el interior de las letras, solo deja el contorno
plt.imshow(gradiente,cmap='gray')
```
