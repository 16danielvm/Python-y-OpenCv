```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
cd "D:/Universidad/Cursos/Python y OpenCv"
```

```python
ajedrez = cv2.imread('tablero-ajedrez.png') #lee la imagen
ajedrez_gris = cv2.cvtColor(ajedrez,cv2.COLOR_BGR2GRAY) #convierte la imagen a escala de grises
plt.imshow(ajedrez_gris,cmap='gray') #muestra la imagen
```

```python
esquinas = cv2.goodFeaturesToTrack(ajedrez_gris, 100,0.01,10) #deteccion de esquinas
```

```python
esquinas = np.int0(esquinas) #Convierte los datos a entero
```

```python
esquinas
```

```python
for i in esquinas:
    x,y = i.ravel() #Para coger 1 a 1
    cv2.circle(ajedrez,(x,y),4,(255,0,0),8)
```
