```python
#Dibujamos figuras en una imagen
```

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
imagen = np.zeros(shape=(500,500,3),dtype=np.int16) #Crea una imagen negra con las dimensiones de (500,500)
```

```python
imagen.shape
```

```python
plt.imshow(imagen)
```

```python
cv2.rectangle(imagen, pt1=(20,20), pt2=(100,100), color=(255,0,0), thickness=10) #Dibuja un rectangulo en (imagen, punto1 (x,y) ,punto2(x,y), color en RGB (R,G,B), grosor de la figura)
```

```python
plt.imshow(imagen)
```

```python
cv2.circle(imagen, center=(250,250), radius=100,color=(0,255,0),thickness=10) #Dibuja un circulo en (imagen, centro (x,y) , radio , color en RGB(R,G,B), grosor de la figura)
plt.imshow(imagen)
```

```python
cv2.line(imagen, pt1=(0,400),pt2=(400,400), color=(0,0,255),thickness=10) #Dibuja un rectangulo en (imagen, punto1 (x,y) ,punto2(x,y), color en RGB (R,G,B), grosor de la figura)
plt.imshow(imagen)
```
