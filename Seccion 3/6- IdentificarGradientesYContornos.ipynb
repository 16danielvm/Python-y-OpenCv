```python
imagen = cv2.imread('sudoku.jpg',0)
plt.imshow(imagen, cmap='gray')
```

```python
sobelx = cv2.Sobel(imagen, cv2.CV_64F, 1, 0, ksize=5) #El paramtro 1,0 captura lineas verticales 
plt.imshow(sobelx, cmap='gray')
```

```python
sobely = cv2.Sobel(imagen, cv2.CV_64F, 0, 1, ksize=5) #Los parametros 0,1 captura lineas horizontales
plt.imshow(sobely, cmap='gray')
```

```python
imagen_2 = cv2.addWeighted(src1=sobelx,alpha=0.5,src2=sobely,beta=0.5,gamma=0)
plt.imshow(imagen_2, cmap='gray')
```

```python
laplacian = cv2.Laplacian(imagen, cv2.CV_64F)
plt.imshow(laplacian,cmap='gray')
```
