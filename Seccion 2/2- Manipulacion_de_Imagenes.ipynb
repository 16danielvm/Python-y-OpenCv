```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
imagen = cv2.imread('perro.jpg')
imagen = cv2.cvtColor(imagen,cv2.COLOR_BGR2RGB) #PLT entiende RGB
```

```python
plt.imshow(imagen)
```

```python
imagen.shape #(y,x,3)
```

```python
imagen_2= cv2.resize(imagen,(600,500)) #Metodo para modificar el tamaño (x,y)
```

```python
plt.imshow(imagen_2)
```

```python
ration_ancho = 0.5
ratio_alto = 0.5
```

```python
imagen_2 = cv2.resize(imagen_2,(0,0),imagen_2,ration_ancho,ratio_alto)
```

```python
plt.imshow(imagen_2)
```

```python
imagen_2.shape #(y,x,3)
```

```python
imagen_2 = cv2.flip(imagen_2,1) #Voltea la imagen de izquierda a derecha
```

```python
plt.imshow(imagen_2)
```

```python
imagen_2 = cv2.flip(imagen_2,0) #Voltea la imagen al reves
```

```python
plt.imshow(imagen_2)
```

```python
imagen = cv2.cvtColor(imagen,cv2.COLOR_RGB2BGR) 
```

```python
cv2.imwrite('perro_al_reves.jpg', imagen_2) #Guarda imagen (Nombre_imagen, imagen)
```
