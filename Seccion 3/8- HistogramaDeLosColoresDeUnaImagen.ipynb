```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
imagen_1 = cv2.imread('perro.jpg')
```

```python
plt.imshow(imagen_1)
```

```python
imagen_2 = cv2.cvtColor(imagen_1,cv2.COLOR_BGR2RGB)
plt.imshow(imagen_2)
```

```python
imagen_2.shape #Muestra (alto, ancho y numero de matrices)
```

```python
histo = cv2.calcHist([imagen_2],[0],None,[256],[0,256]) #0-> R,None,Valor Maximo, [Rango de Valores]
plt.plot(histo)
```

```python
#Bucle para ver los tres colores
colores = ('b','g','r')
for i,color in enumerate(colores):
    histo = cv2.calcHist([imagen_2],[i],None,[256],[0,256])
    plt.plot(histo)
    plt.xlim([0,256])
```
