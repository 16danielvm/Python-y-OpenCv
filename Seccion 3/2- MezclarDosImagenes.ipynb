```python
imagen_1 = cv2.imread('perro.jpg')
```

```python
imagen_2 = cv2.imread('marca-agua-redaitcurso.jpg')
```

```python
imagen_2 = cv2.cvtColor(imagen_2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(imagen_2)
```

```python
imagen_1 = cv2.resize(imagen_1,(1200,1200)) #Modifica el tamaño de la imagen
```

```python
imagen_2 = cv2.resize(imagen_2,(1200,1200)) #Modifica el tamaño de la imagen
```

```python
plt.imshow(imagen_1)
```

```python
plt.imshow(imagen_2)
```

```python
imagen_mezcla = cv2.addWeighted(src1=imagen_1, alpha=0.8,src2=imagen_2, beta=0.2,gamma=0) #Funcion para mezclar dos imagenes fuente y peso de las imagenes
```

```python
plt.imshow(imagen_mezcla)
```
