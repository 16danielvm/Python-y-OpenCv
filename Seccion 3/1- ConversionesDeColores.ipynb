```python
imagen = cv2.imread('perro.jpg') #Cv2 lo lee en formato BGR(Blue Green Red)
plt.imshow(imagen) #Plt utiliza RGB (Red Green Blue)
```

```python
#Cambiamos la escala de colores
imagen_2 = cv2.cvtColor(imagen, cv2.COLOR_BGR2RGB) #Cambia de BGR a RGB
plt.imshow(imagen_2)
```

```python
imagen_2 = cv2.cvtColor(imagen, cv2.COLOR_BGR2HSV) #Cambia de BGR a HSV
plt.imshow(imagen_2)
```

```python
imagen_2 = cv2.cvtColor(imagen, cv2.COLOR_BGR2HLS)
plt.imshow(imagen_2)
```
