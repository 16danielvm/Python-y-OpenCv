```python
imagen = cv2.imread('perro.jpg',0) #Adicionando el 0 indica que lea la imagen en escala de grises
plt.imshow(imagen, cmap='gray')
```

```python
mitad, imagen = cv2.threshold(imagen, 255/2, 255, cv2.THRESH_BINARY) #Mitad: lo que considera , imagen: lo que devuelve
plt.imshow(imagen, cmap='gray')
```

```python
mitad, imagen = cv2.threshold(imagen, 255/2, 255, cv2.THRESH_BINARY_INV) #Realiza el proceso inverso lo blanco lo pone negro y lo negro lo pone blanco
plt.imshow(imagen, cmap='gray')
```
